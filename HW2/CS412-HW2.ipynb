**2.a)** What is the probability that a respondent is currently employed as a Programmer given they use C/C++ at work?

```python
def bayes_theorem(pA, pBA, pB):
  return (pBA * pA) / pB
```

```python
# P(Programmer | C/C++) = P(C/C++ | Programmer) * P(Programmer) / P(C/C++)

probability_of_programmer = (df['CurrentJobTitleSelect'] == 'Programmer').sum() / len(df)

print("P(Programmer) =", probability_of_programmer)



probability_of_CPP = df['WorkToolsSelect'].str.contains("C\/C\+\+", case=False, na=False).sum() / len(df)


print("P(C/C++) =", probability_of_CPP)

# P(C/C++ | Programmer) = P(C/C++  Programmer) / P(Programmer)
# P(A|B) = P(AB) / P(B)

probability_of_CPP_given_programmer = ((df['CurrentJobTitleSelect'] == 'Programmer') & (df['WorkToolsSelect'].str.contains("C\/C\+\+", case=False, na=False))).sum() / ((df['CurrentJobTitleSelect'] == 'Programmer')).sum()

print("P(C/C++ | Programmer) =", probability_of_CPP_given_programmer)


result = bayes_theorem(probability_of_CPP_given_programmer, probability_of_programmer, probability_of_CPP)

print("P(Programmer | C/C++) =", result)

print("The result is", result)
```

**2.b)** What is the probability that a respondent is a Data Scientist given they have majored in computer science, mathematics or statistics?

```python
# P(Data Scientist | Computer Science U Mathemathics U Statistics) = P(Computer Science U Mathemathics U Statistics | Data Scientist) * P(Data Scientist) / P(Computer Science U Mathemathics U Statistics)


prob_data_scientist = (df['CurrentJobTitleSelect'] == 'Data Scientist').sum() / len(df)

print("P(Data Scientist) = ",prob_data_scientist)

prob_cs_math_stat = ((df['MajorSelect'] == 'Computer Science') | (df['MajorSelect'] == 'Mathematics or statistics')).sum() / len(df)

print("P(Computer Science U Mathemathics U Statistics) =", prob_cs_math_stat)


# P(Computer Science U Mathemathics U Statistics | Data Scientist) = P((Computer Science U Mathemathics U Statistics)  Data Scientist) / P(Data Scientist)
# P(A|B) = P(AB) / P(B)


prob_cs_math_stat_given_data_scientist = ((df['CurrentJobTitleSelect'] == 'Data Scientist') & ((df['MajorSelect'] == 'Computer Science') | (df['MajorSelect'] == 'Mathematics or statistics'))).sum() / (df['CurrentJobTitleSelect'] == 'Data Scientist').sum()

print("P(Computer Science U Mathemathics U Statistics | Data Scientist) =", prob_cs_math_stat_given_data_scientist)


result = bayes_theorem(prob_cs_math_stat_given_data_scientist, prob_data_scientist, prob_cs_math_stat)

print("P(Data Scientist | Computer Science U Mathemathics U Statistics) =", result)

print("The result is", result)

```

**2.c)** What is the probability that a respondent works in the Technology industry given that they earn more than 40,000 USD annually?


```python
# P(Technology | Salary > 40.000 USD) = P(Salary > 40.000 USD | Technology) * P(Technology) / P(Salary > 40.000 USD)


# P(Technology)
prob_tech = (df['EmployerIndustry'] == 'Technology').sum() / len(df)
print("P(Technology) =", prob_tech)



# P(Salary > 40.000 USD)
# df['CompensationAmount'] = pd.to_numeric(df['CompensationAmount'], errors="coerce")
prob_salary = (df['CompensationAmountUSD'] > 40000).sum() / len(df)
print("P(Salary > 40.000 USD) =", prob_salary)


# P(Salary > 40.000 USD | Technology) = P(Salary > 40.000 USD  Technology) / P(Technology)
prob_salary_given_tech = ((df['EmployerIndustry'] == 'Technology') & (df['CompensationAmountUSD'] > 40000)).sum() / (df['EmployerIndustry'] == 'Technology').sum()
print("P(Salary > 40.000 USD | Technology) =", prob_salary_given_tech)

result = bayes_theorem(prob_salary_given_tech, prob_tech, prob_salary)

print("P(Technology | Salary > 40.000 USD) =", result)

print("The result is", result)
```
